### Gravity
Okay, so the first thing we need to think about is gravity.  We know that on Earth, the acceleration due to gravity is about 9.8 m/s$^2$ -- but how do we calculate this for other bodies of mass?


Well, first let's think about the standard equation for force, 

$$F=m_1a$$

In this case, $a$ is the acceleration due to gravity so we can also represent it as $a_g$.  This makes that equation

$$F_g=m_1a_g$$

#### Interesting!
Now this still points back to our main problem of not knowing what $a_g$ is for other planets/large bodies.  But there's an equation that we can leverage for this.  Look up the equation for **Newtonâ€™s Law of Universal Gravitation** and write it down.

Notice that this equation is also one that can tell you the force of gravity, $F_g$.  

**How can you use both equations to solve for $a_g$?**

```python
# write out the final equation as a function
# where gravity is a_g  (let me know if you get stuck)


# think about what input values you need to solve for this
def get_gravity(m_co, r):
    G = 6.67430E-11
    #m_co = int(input("Enter the mass of the celestial object: "))
    #r = int(input("Enter the radius of the celestial object: "))
    a_g = (G*m_co)/r**2
    return a_g    
    #print (gravity, a_g)
get_gravity(5.972E24,6378e3)
```

## Writing this in code
Alright, now that you have the equations that you need, let's code it!

```python
# first, let's set up our given values
x0 = 0       # meters
y0 = 1.7      # meters (feel free to use your own height here!)
v = 5      # m/s
ang = np.radians(40.)      # radians (hint: check out np.radians() :)
# now for gravity
ea_g = -9.8 # m/s^2, negative because it's pulling down
```

```python
# making an array for time, t
t = np.linspace(0,50,10000) # so we can plug in time to the x & y equations
```

```python
# next, calculate your velocity components
vx = np.cos(ang)* v

vy = np.sin(ang)* v
```

```python
# now, write the equations for x and y using all of these values

x = x0 + vx*t
y = y0 + vy +(.5*ea_g*(t**2))
```

### Sweet!  Let's look at it.
Now we're going to plot these equations to see the path of the projectile.  Feel free to play with the plotting parameters, adding labels or having fun, etc.

```python
x = x[y>=0]
y = y[y>=0]
```

```python
plt.figure(figsize=(9,4))

plt.plot(x,y)
plt.title('x vs. y Graph')
plt.xlabel('Horizontal Distance Traveled (m)')
plt.ylabel('Vertical Distance Traveled (m)')

plt.tight_layout()
plt.show()
plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig = plt.figure(figsize = (9.5,5))
ax = plt.gca()

t = np.linspace(0,50,10000)
ang = np.radians(40.)
a_g = 133391337083.90112

v = 3e11
vx = np.cos(ang)* v
vy = np.sin(ang)* v

x0 = 0
y0 = 1.7
x = x0 + vx*t
y = y0 + vy*t -(.5*a_g*(t**2))
x = x[y>=0]
y = y[y>=0]

ax.set_title('Throwing a ball on a Neutron Star at 300 times the speed of light')
ax.set_xlabel('distance [m]')
ax.set_ylabel('height [m]')
axe, = ax.plot(x[0], y[0])
ball = plt.Circle((x[0], y[0]), radius=1e10, fc='black')
#exa = plt.axes(xlim=(-0.2,70), ylim=(-0.2,70)) 

# need an initial setup function so the animation function can
# anchor off of it -- this is essential!
def init():
    axe.set_data(x[0],y[0])
    ball.center =(x[0], y[0])
    ax.add_patch(ball)
    return axe, ball,

# the function that will actually be changing things in the animation
def throw(i):
    #x, y = ball.center
    axe.set_data(x[:i],y[:i])
    ball.center= (x[i-1],y[i-1])
    ax.set_ylim(-0.2,max(y[:i])+max(y[:i])*0.4)
    ax.set_xlim(-0.2,max(x[:i])+max(x[:i])*0.2)
    return axe, ball,

# all the build up lead to this! Running the animation function...
anim = animation.FuncAnimation(fig,throw,init_func=init,\
    frames=np.arange(1,len(x)),interval=100,blit=True)

plt.tight_layout()
anim.save('throws.gif', fps=30, writer='imagemagick') # this will take a while to make
#plt.show()
plt.close('all')
```
